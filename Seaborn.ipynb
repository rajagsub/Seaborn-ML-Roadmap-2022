## Scatter Plot
- A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

- relplot[Fig level function]
- scatterplot [Axes level Functions]

```python
data = sns.load_dataset('tips') #--> Pre defined dataset.
data.head()
```

###  plot a scatter plot btw total_bill and tip

```python
 sns.regplot(x='total_bill',y='tip',data=data)
```

### The Hue Plot - categorical data

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

### Style parameter - categorical data

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

### Size parameter - continious data or numerical data

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size', sizes=(15,200),data=data)
```

### Scatter Plot function

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
```

## Categorical Scatterplots | Stripplots and Swarmplots

```python
# x-axis --> Categorical data
# y-axis --> Numberical data
```

```python
tips= sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=0.10,data=tips)

#jitter -> used to create a noice in data to show scatterly 
```

## Swam Plot - shows some distribution of the data

```python
sns.catplot(x='day',y='tip',kind='swarm',data=tips)
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='swarm',hue='sex')
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
```

## Box Plots

The image above is a boxplot. A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.




```python
# Univariate analyis - this will give us the only one box plot for single field.
sns.boxplot(tips['total_bill'])
```

```python
# Now lets use the catplot and use multiple variable to do multivariate analysis
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='box',data=tips)
```

## Violinplots - 
Plotting numberical data and can be considered a combination of the box plot with the kernel density plot



the advantage of the violin plot over box is it shows the stats and it also shows the entier distribution of the data

```python
# Univariate
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex', split=True,data=tips)
```

## Barplots and Countplots

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=tips,hue='sex')
```

```python
sns.catplot(x='smoker',y='total_bill',estimator= np.std,kind='bar',data=tips,hue='sex')
```

```python
#count plots - spl case of bar plot - similar to histogram
sns.catplot(x='sex',kind='count',data=tips,hue='smoker')
```

## Heatmaps - categorical data

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
x = flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False, linewidth=0.5, annot=True, fmt='d')
```

```python
sns.heatmap(x,cbar=False, linewidth=0.5, annot=True, fmt='d', cmap='summer')
```

### Cluster Map

```python
sns.clustermap(x,linewidth=0.5, z_score=0 ,
               annot= True,
               row_cluster=False,
               metric='correlation')
```

## Join plot  - Multivariate analysis on numerical data

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#hexplot - 2d histogram

sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips)
```

```python
# 2d KDE plot
sns.jointplot(x='total_bill',y='tip',kind='kde',data=tips,color='green')
```

```python
# Regression Plot

sns.jointplot(x='total_bill',y='tip',kind='reg',data=tips)
```

## Distplots

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

### Displot

```python
sns.distplot(titanic['age'])
# KDE & Histogram together
```

```python
# Plotting Histogram in distplot

sns.distplot(titanic['age'],bins=10,kde=False)
```

```python
sns.distplot(titanic['age'],hist=False)
```

```python
#rugplot
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

```python
#distplot

sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
